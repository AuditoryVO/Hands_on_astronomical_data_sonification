```python
import csv
import numpy as np
```

```python
with open("Bastien-CDS-table.csv", newline="") as file:
    reader = csv.reader(file)
    
    for row in reader:
        print(row)  # each row is a list
```

```python
# Alternative dictionary style
with open("Bastien-CDS-table.csv", newline="") as file:
    reader2 = csv.DictReader(file)
    for row in reader2:
        print(row["KIC"])
```

```python
# Exporting to CSV
column_index = 3
with open("Bastien-CDS-table.csv", newline="") as file:
    reader = csv.reader(file)    
    column = [row[column_index] for row in reader]
print(column)
with open("KIC_IDs.csv", "w", newline="") as file:
    writer = csv.writer(file)
    for value in column:
        writer.writerow([value])
```
