```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sine_wave_generator(amplitude, freq, phase, duration, sample_rate):
    t = np.linspace(0,duration, int(sample_rate*duration), endpoint=False)
    return t, amplitude * np.sin(2 * np.pi * freq * t + phase)
```

```python
frequency = 100
```

```python
x_axis, sine_wave= sine_wave_generator(1, frequency, 0, 1, 48000)
```

```python
sine_wave
```

```python
x_axis
```

```python
fig, ax = plt.subplots()
plt.axis([0.5, 0.55, -1.2, 1.2])
fig.set_size_inches(12., 8.)
ax.plot(x_axis, sine_wave, 'r')
fig.suptitle(f"{frequency}Hz Sinewave")
ax.set_ylabel("Normalized amplitude")
ax.set_xlabel("Time (s)")
```
