## Normalization

```python
# min-max Normalization = x-min/max-min
min_mag = min(magnitude_ok)
max_mag = max(magnitude_ok)
print(f"min={min_mag} max={max_mag}")

magnitude_norm = (magnitude_ok-min_mag)/(max_mag-min_mag)
magnitude_norm
```

```python
# Rescaling between a set of values, for instance MIDI values [0-127] = a + (x-min)(b-a)/(max-min)
a= 0
b= 127
magnitude_midi = (a + (magnitude_ok-min_mag)*(b-a)/(max_mag-min_mag))
magnitude_midi = magnitude_midi.astype(int) #Converting to float values into integers
magnitude_midi
```

```python
# Mean normalization = x-mean/max-min
mean_mag = np.mean(magnitude_ok)
print(f"mean={mean_mag}")

magnitude_mean_norm = (magnitude_ok-mean_mag)/(max_mag-min_mag)
magnitude_mean_norm
```

```python
# Standardization(Z-score normalization) = x-mean/std

std_mag = np.std(magnitude_ok)
print(f"standard deviation={std_mag}")

magnitude_std_norm = (magnitude_ok-mean_mag)/std_mag
magnitude_std_norm
```
