```python
import rtmidi
```

```python
# Basic commands to send MIDI values to an external Synth or DAW with rtmidi
midiout = rtmidi.MidiOut()
available_ports = midiout.get_ports()

if available_ports:
    midiout.open_port(0)  # Open first available port
else:
    midiout.open_virtual_port("Virtual Output")
 
```

```python
note = 60        # C4 => f=440⋅2^((n−69)/12) with n the MIDI value
amplitude = 127  # Maximum amplitude => 128 MIDI amplitude values (0-127)

note_on = [0x90, note, amplitude] # Channel 1, 0x90 <=> Channel 2, 0x91 <=> Channel 3, ...
midiout.send_message(note_on)
```

```python
note = 60        # C4 => f=440⋅2^((n−69)/12) with n the MIDI value
amplitude = 127  # Maximum amplitude => 128 MIDI amplitude values (0-127)

note_on = [0x91, note, amplitude] # Channel 1, 0x90 <=> Channel 2, 0x91 <=> Channel 3, ...
midiout.send_message(note_on)
```

```python
note_off = [0x80, note, amplitude]  # Note off on same Channel 1, 0x81 <=> Channel 2, 0x82 <=> Channel 3, ...
midiout.send_message(note_off)
```

```python
note_off = [0x81, note, amplitude]  # Note off on same Channel 1, 0x81 <=> Channel 2, 0x82 <=> Channel 3, ...
midiout.send_message(note_off)
```

```python
# alternative method to note off
note_on = [0x90, note, 0]
midiout.send_message(note_on)
```
