```python
from pythonosc import udp_client
```

```python
# Set the IP and Port
# "127.0.0.1" is localhost; 5005 is a common default port
ip = "127.0.0.1"
port = 5005

# Initialize the client
client = udp_client.SimpleUDPClient(ip, port)
```

```python
# Send a single frequency (float) value
freq = 500
client.send_message("/freq", freq)

print(f"Frequency = {freq}Hz sent to {ip} : {port}")
```

## Multiple ports

```python
#play
amp = 0.5
client1 = udp_client.SimpleUDPClient("127.0.0.1", 9991) 
client1.send_message("/amp", amp)
print("Amplitude =", amp)
```

```python
#stop via amp
client1.send_message("/amp", 0)
print("Amplitude =", 0)
```

```python
#stop via frequency
client.send_message("/freq", 0)
print("Frequency =", 0)
```
