## Project 3. Sonification of potential asteroids detected by the Hubble Space Telescope.

Design and develop a sonification project to represent potential asteroids detected by the Hubble Space Telescope. You can implement your own synthesizers and/or use any virtual instrument. You can also choose to control them by MIDI or OSC. In the following steps you have a mapping strategy proposal but you can design your own approach. 

About asteroids: https://science.nasa.gov/solar-system/asteroids/
The catalog: https://cdsarc.cds.unistra.fr/viz-bin/cat/J/A+A/659/A38
Racero 2022: https://www.aanda.org/articles/aa/full_html/2022/03/aa40899-21/aa40899-21.html

- Use Aladin Sky Atlas to obtain the Catalog of potential detections of asteroids (Racero+, 2022) and visualize the Catalog with the list of potential detections from the Hubble Space Telescope.
- Export the table to a .txt file and clean the data.
- Map the magnitude (Vmag1) to the pitch or fundamental frequency of each note to represent the asteroid.
- Use the actual coordinates of the objects (RA1/DE1, RA2/DE2) to provide a representation of the trajectory of the asteroid within a binaural soundscape.
- Explore the catalog sequentially. Map the duration of the observation (Dur) to the duration of the sound or note.


1. Import the dependencies

2. Load the file "Asteroids_CDS-J-A+A-659-A38-hstaster.txt" as a pandas dataframe. You may need to clean the dataset.

3. Extract the columns for the representation (Vmag1, RA1, DE1, RA2, DE2)

4. Convert all the lists to numpy arrays

5. Scale the Magnitude values and convert them into integers to fit an audible range between 100 and 1000 Hz. 

6. Convert the duration to rounded integers

7. Find the quartiles of the duration values and use them to set the thresholds that allow the conversion of the durations into 1, 1/2, 1/4 and 1/8 notes.

8. Convert the "durations" array into a "notes" array with the following values:
   - 0.25 for the durations under the first quartile
   - 0.5 for the durations between de first and the second quartiles
   - 0.75 for the durations between the second and the third quartiles
   - 1 for the durations over the third quartile

9. Define a function to represent the trajectory of each asteroid from (RA1, DE1) to (RA2, DE2) acording to the duration of the observation.

10. Create a loop to send the OSC messages to your CSound synthesizer. Remember that CSound only receives positive values, so take the absolute value of data<0 before sending and recover the negative value at the CSound receiver.

11. Generate the sonification implementing your own Synth in Cabbage/Csound
